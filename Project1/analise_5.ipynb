## Criando novas variáveis

```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('dados/aluguel_residencial_2.csv', sep=';')
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = (dados['Valor'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
list = dados.Tipo.unique()
```

```python
list = sorted(list, key = lambda i: i[0])
list
```

```python
casa =  ['Casa de Condomínio', 'Casa', 'Casa Comercial', 'Casa de Vila',]
```

```python
dados ['Tipo Agregado'] =  dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(30)
```

### Excluindo Variaveis
